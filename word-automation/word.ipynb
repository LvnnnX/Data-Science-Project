```python
def change_value(docs, num_tables, start_baris, kolom, num, **kwargs):
    # print(kwargs.items())
    if('values' in kwargs.keys()):
        values = kwargs.get('values')
        print(f'values found, the values is {kwargs.get("values")}')
    else:
        values = f'Default value {num}'


    for _ in range(100):
        format = {'num':num, 'values':values, 'num_tables':num_tables, 'start_baris':start_baris, 'kolom':kolom}
        try:
            docs.tables[num_tables].cell(start_baris,kolom).text = values.format(**format)
            docs.tables[num_tables].cell(start_baris,kolom).paragraphs[0].runs[0].font.bold = True
            docs.tables[num_tables].cell(start_baris,kolom).paragraphs[0].runs[0].font.size = Pt(12)
            docs.tables[num_tables].cell(start_baris,kolom).paragraphs[0].runs[0].aligment = WD_ALIGN_PARAGRAPH.CENTER
            docs.tables[num_tables].cell(start_baris,kolom).paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
            docs.tables[num_tables].cell(start_baris,kolom).paragraphs[0].runs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
            start_baris+=1
            num+=1
        except:
            break
    return docs
```

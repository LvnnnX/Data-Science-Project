```python
import pathlib
import pandas as pd
import numpy as np

BASE_DIR = pathlib.Path().parent
BOOK_DIR = BASE_DIR / "buku1"
SHEET_DIR = BOOK_DIR / "ch6"

book_name = 'bab6_adult_data.csv'
```

Read Data

```python
df = pd.read_csv(f'{SHEET_DIR}/{book_name}')
df.head(5)
```

Cek Data Null

```python
df.isnull().sum()
df.dropna(inplace=True)
```

Mencari features yang berkorelasi dengan data target

```python
import sklearn.preprocessing as prep
#Mengubah data sex(jenis kelamin) kedalam data numerik
gender = prep.LabelEncoder()
gender.fit(df['sex'])
df['sex_code'] = gender.transform(df['sex'])
getsex = df.groupby('sex')[['sex_code']].count()
getsex['sex_to_code'] = np.arange(0,len(getsex))
# df[['sex','sex_code']].head(5)
# Male = 1 ; Female = 0
#End data sex changes 
getsex
```

```python
#Mengubah data education(edukasi) kedalam data numerik
edu = prep.LabelEncoder()
edu.fit(df['education'])
df['edu_code'] = edu.transform(df['education'])
getedu = df.groupby('education')[['edu_code']].agg('count')
getedu['education_to_num'] = np.arange(0,len(getedu))
# df[['edu_code','education']].sort_values(by='edu_code')
getedu
```

```python
#Mengubah data race(ras) kedalam data numerik
race = prep.LabelEncoder()
race.fit(df['race'])
df['race_code'] = race.transform(df['race'])
getrace = df.groupby('race')[['race_code']].agg('count')
getrace['race_to_num'] = np.arange(0,len(getrace))
# df[['race_code','race']].sort_values(by='race_code')
getrace
```

```python
#Mengubah data workclass(kelas pekerja) kedalam data numerik
wcl = prep.LabelEncoder()
wcl.fit(df['workclass'])
df['wcl_code'] = wcl.transform(df['workclass'])
getwcl = df.groupby('workclass')[['wcl_code']].agg('count')
getwcl['workclass_to_num'] = np.arange(0,len(getwcl))
# df[['wcl_code','workclass']].sort_values(by='wcl_code')
getwcl
```

```python
#Mengubah data occupation(pekerjaan) kedalam data numerik
ocp = prep.LabelEncoder()
ocp.fit(df['occupation'])
df['ocp_code'] = ocp.transform(df['occupation'])
getocp = df.groupby('occupation')[['ocp_code']].agg('count')
getocp['occupation_to_num'] = np.arange(0,len(getocp))
# df[['ocp_code','occupation']].sort_values(by='ocp_code')
getocp
```

```python
#Mengubah data relationship(hubungan) kedalam data numerik
rlts = prep.LabelEncoder()
rlts.fit(df['relationship'])
df['rlts_code'] = rlts.transform(df['relationship'])
getrlts = df.groupby('relationship')[['rlts_code']].agg('count')
getrlts['relationship_to_num'] = np.arange(0,len(getrlts))
# df[['rlts_code','relationship']].sort_values(by='rlts_code')
getrlts
```

```python
#Mengubah data native_country(?) kedalam data numerik
ntv = prep.LabelEncoder()
ntv.fit(df['native_country'])
df['ntv_code'] = ntv.transform(df['native_country'])
getntv = df.groupby('native_country')[['ntv_code']].agg('count')
getntv['native-country_to_num'] = np.arange(0,len(getntv))
# df[['ntv_code','native_country']].sort_values(by='ntv_code')
getntv
```

Membuang semua features yang sudah menjadi kode

```python
X = df.drop(['fnlwft','workclass','education','relationship','native_country','occupation','race','sex','marital_status','income','capital_gain','capital_loss','sex_code','race_code'],axis=1)
y = df['income']
X
```

Setelah mendapatkan data numerik, sekarang buat Random Forest

```python
import sklearn.model_selection as ms
import sklearn.tree as tree
import sklearn.ensemble as ens
import sklearn.metrics as met

rf = ens.RandomForestClassifier(n_estimators=100)

X_train,X_test,y_train,y_test = ms.train_test_split(X,y,test_size=0.2)
rf.fit(X_train,y_train)

y_predict = rf.predict(X_test)
print(f'Akurasi: {met.accuracy_score(y_test,y_predict)}')
```

```python
cek=rf.feature_importances_
for x in cek:
    print(x*100)
```
